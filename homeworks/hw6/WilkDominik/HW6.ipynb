### Analiza zmiennych numerycznych w zależności od przynależności do grupy

```python
df_type = df
df_type['genres'] = df_type['genres'].str.split(', ')
df_exploded = df_type.explode('genres')
```

```python
from IPython.display import display, Markdown
avg_table1 = df_exploded.groupby('genres')['budget'].describe().sort_values(by = 'mean', ascending=False)
avg_table2 = df_exploded.groupby('genres')['revenue'].describe().sort_values(by = 'mean', ascending=False)
display(Markdown("####  Budżet według gatunku"))
display(avg_table1)
display(Markdown("#### Przychód według gatunku"))
display(avg_table2)
```

### Wnioski
Widzimy że zarówno w tabeli dotyczącej budżetu i przychodu na pierwszych 4 miejscach są te same gatunki filmów czyli: Przygodowy, Rodzinny, Fantastyczny i Komediowy. Oznacza to że filmy z tych gatunków zarówno wymagają największego nakłądu pieniedzy ale przynosza też największe zyski.

```python
from itertools import combinations
from collections import Counter
genre_pairs = []

for genres in df['genres']:
    unique_genres = set(genres)  
    unique_genres.remove("Animation") # usuwamy Animation bo dane zawierają tylko filmy animowane więc ten gatunek występuje w każdym filmie
    pairs = combinations(sorted(unique_genres), 2) 
    genre_pairs.extend(pairs)

pair_counts = Counter(genre_pairs)
```

```python
top_pairs = pair_counts.most_common(10)

# Dane do wykresu
labels = [f"{a} & {b}" for (a, b), _ in top_pairs]
values = [count for _, count in top_pairs]

# Wykres
plt.figure(figsize=(10, 6))
plt.barh(labels, values, color='skyblue')
plt.xlabel("Liczba wspólnych wystąpień")
plt.title("Top 10 najczęstszych połączeń gatunków filmowych")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

### Wnioski
Najczęstsze połączenie gatunków w filmach to Komedia i Rodzinny, Przygodowy i Rodzinny oraz Fantastyczny i Rodzinny. Zatem widzimy że bardzo duża część filmów animowanych to filmy należace do gatunku filmów rodzinnych

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['budget'], df['revenue'], alpha=0.6, color='blue')
plt.title('Zależność Przychodu od Budżetu')
plt.xlabel('Budżet')
plt.ylabel('Przychód')
plt.grid(True)
plt.show()
```

### Wnioski
Widzimy że wraz ze wzrostem budżetu filmy osiągają coraz większy przychód.

```python
filtered_df = df[
    (df['runtime'] <= df['runtime'].quantile(0.98)) &  
    (df['runtime'] > 0 ) &
    (df['budget'] > 0)                                  
]
filtered_df.dropna()
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['runtime'], filtered_df['budget'], alpha=0.6, color='green')
plt.title('Zależność budżetu od długości filmu')
plt.xlabel('Długość filmu')
plt.ylabel('Budżet')
plt.grid(True)
plt.show()
```

### Wnioski
Widzimy że wraz ze wzrostem długości filmu budżet wymagany do stworzenia takiego filmu wzrasta.

## Zadanie 4
### Jeden wykres analizujący rozkład wartości wybranej zmiennej jakościowej typu multilabel (czyli takiej, która zawiera więcej niż jedną wartość dla pojedynczej obserwacji, np. zmienne opisujące gatunki gier/filmów), wraz z kluczowymi wnioskami

```python
df_temp = df.dropna(subset=["runtime", "genres"]).copy()
df_temp["genres"] = df_temp["genres"].apply(lambda x: [g.strip() for g in x.split(",")])
temp = df_temp.explode("genres")
temp2 = temp.groupby("genres")["runtime"].mean().sort_values(ascending=False)
print(temp2.head(10))
```

```python
plt.figure(figsize=(10, 6))
temp2.plot(kind="bar", color="pink", edgecolor="black")

plt.title("Średni czas trwania filmu w zależności od jego gatunku", fontsize=14)
plt.xlabel("Gatunek", fontsize=12)
plt.ylabel("Średni czas trwania filmu (minuty)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

Średni czas trwania filmu jest najdłuższy dla gatunków 'Crime', 'Adventure', 'Action' i 'Mystery'; dla tych gatunków przekrzacza on próg 50 minut. Natomiast najkrótszą średnią długość mają filmy z gatunku 'Animation', 'Documentary' i 'Western' i wynosi ona poniżej 25 minut.

## Zadanie 5
### Dwa wykresy wskazujące zależności pomiędzy wybranymi zmiennymi w zbiorze danych, wraz z kluczowymi wnioskami.

```python
df_plot1 = df.dropna(subset=["popularity", "vote_average"])
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_plot1, x="popularity", y="vote_average", alpha=0.6)
plt.title("Popularność a średnia oceną")
plt.xlabel("Popularność")
plt.ylabel("Średnia ocena")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

Wnoski: Popularność filmów rośnie wraz z ich oceną; największą popularność mają filmy o ocenie 7 i powyżej.

```python
top_countries = df['production_countries'].value_counts().index[:10]
df_top = df[df['production_countries'].isin(top_countries)]

plt.figure(figsize=(14, 8))
sns.violinplot(data=df_top, x='production_countries', y='runtime')
plt.title("Długości filmów według krajów produkcji ")
plt.xlabel("Kraj produkcji")
plt.ylabel("Długość filmu")
plt.xticks(rotation=45)
plt.show()
```

Zauważmy, że spośród 10 krajów, które najczęściej pojawiają się jako kraj produkcji, krajami zdecydowanie wyróżniającymi się pod względem długości produkowanych filmów są USA, Japonia oraz Francja. Można również zauważyć, że zarówno Rosja, jak i Niemcy i Chiny mają spory odsetek filmów o długości około 160-180 minut. W pozostałych krajach wielkości te są sporo niższe.

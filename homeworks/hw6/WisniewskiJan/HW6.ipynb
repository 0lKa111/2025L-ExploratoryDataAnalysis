Gry, które były darmowe mają zdecydowanie gorsze oceny. Ponadto można w nich zdobyć prawie dwa razy mniej osiągnięć niż w płtanych

```python
all_genres = df['genres'].dropna().str.split(', ')
flat = all_genres.explode()
genre_counts = flat.value_counts().head(10)
```

```python
genres = genre_counts.index
counts = genre_counts.values
plt.bar(genres, counts)
plt.title('Top 10 najpopularniejszych gatunków gier')
plt.ylabel('Liczba gier')
plt.xlabel('Gatunek')
plt.xticks(rotation=45, ha='right')

plt.show()
```

Najpopularniejszym gatunkiem gier są gry akcji, których jest prawie 30000, najmniej jest gier RPG

```python
plt.scatter(df['price_initial (USD)'], df['positive_percentual'], alpha=0.5)
plt.title('Cena początkowa vs. % pozytywnych recenzji')
plt.xlabel('Cena początkowa (USD)')
plt.ylabel('% pozytywnych recenzji')
plt.show()
```

Najdroższa gra nie miała pozytywnych opinii

```python
max_price = df['price_initial (USD)'].max()
df_filtered = df[df['price_initial (USD)'] < max_price]
plt.scatter(df_filtered['price_initial (USD)'], df_filtered['positive_percentual'], alpha=0.5)
plt.title('Cena początkowa vs. % pozytywnych recenzji (bez najdroższej gry)')
plt.xlabel('Cena początkowa (USD)')
plt.ylabel('% pozytywnych recenzji')
plt.show()
```

Dla gier których początkowa cena była mniejsza od 50$ ciężko zauważyc reacje między ceną, a procentem pozytywnych opinii. Dla gier o cenie 200 dolarów opinie są podzielone, zaś dla jeszcze droższych gier procent pozytywnych opini jest wysoki

```python
plt.scatter(df['n_achievements'], df['positive_percentual'], alpha=0.5)
plt.title('Osiągnięcia vs. % pozytywnych recenzji')
plt.xlabel('Liczba osiągnięć')
plt.ylabel('% pozytywnych recenzji')
plt.show()
```

Gry, w któych liczba osiągnięć przekracza 2000 jest więcej pozytywnych niż negatywnych ocen

```python
df_m = df[df['n_achievements'] <= 2000]
plt.scatter(df_m['n_achievements'], df_m['positive_percentual'], alpha=0.5)
plt.title('Osiągnięcia vs. % pozytywnych recenzji (osiągnięcia ≤ 2000)')
plt.xlabel('Liczba osiągnięć')
plt.ylabel('% pozytywnych recenzji')
plt.tight_layout()
plt.show()
```

```python
Dla gier, w których liczba osiągnięć jest mniejsza niz 2000 jest więcej tych któych procent ocen pozytywnych jest wyższy niż 50% lecz dla gier o liczbie osiągnięć większej niż 2
```

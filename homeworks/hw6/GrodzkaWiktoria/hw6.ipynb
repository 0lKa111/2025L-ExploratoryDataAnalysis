## 6.Wykres zmiennej multilabel

```python
df['genres'] = df['genres'].apply(lambda x: ','.join([g.strip() for g in x.split(',')]) if pd.notnull(x) else x)

genres_dummies = df['genres'].str.get_dummies(sep=',')
genre_revenue = {}
for genre in genres_dummies.columns:
    genre_revenue[genre] = df[genres_dummies[genre] == 1]['revenue'].mean()

pd.Series(genre_revenue).sort_values(ascending=False).plot(kind='bar', figsize=(10,5), title='Średni przychód wg gatunku')
plt.ylabel('Przychód')
plt.xlabel('Gatunek')
plt.show()

```

### Wnioski
Jak widzimy największy przychód dają filmy przygodowe.

## 7. Pozostałe wykresy
    


```python
sns.set(rc = {'figure.figsize': (12,7)})
sns.set_style("darkgrid")
pd.options.display.max_rows = 10
pd.set_option('display.width', 5000)
pd.set_option('display.max_columns', 60)
```

### Wykres 1 

```python
sns.scatterplot(data=df, x='budget', y='revenue', size='popularity')
plt.ylabel('Przychód')
plt.xlabel('Budżet')
plt.legend(title='Popularność')

plt.show()
```

### Wnioski
Jak widać na wykresie, duży dudżet nie zapewnia wysokich przychodów. Dodatkowo film nie musi być popularny by zwracać wysoki przychód.

### Wykres 2

```python
sns.scatterplot(data=df, x='vote_average', y='revenue', hue='vote_count')
plt.ylabel('Przychód')
plt.xlabel('Średnia ocena')
plt.legend(title='Liczba oddanych głosów')

plt.show()
```

### Wnioski
Jak widzimy na wykresie, jeżeli ocena widzów jest większa niż 5 i liczba głośów jest większa niż 4000 to jest szansa na wysokie zyski. Co ciekawe, filmy o ocenie powyżej 9 osiągają znacznie mniejsze przychody, jednak może to byś spowodowane małą liczbą oddanych głosów.

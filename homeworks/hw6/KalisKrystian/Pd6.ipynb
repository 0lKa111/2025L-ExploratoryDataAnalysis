```python
gr = data.groupby('is_free')[['n_achievements',
    'total_reviews', "positive_percentual"]].agg(['mean', 'std'])
print(gr)
```

Darmowe gry mają dwa około dwa razy więcej osiągnięć. Dla darmowych gier odchylenie standardowe 
całkowitej liczby recenzji jest dwa razy mniejsze niż dla gier płatnych. Płatne gry mają lepsze recenzje.

**4**

```python
genres_raw = data['genres'].fillna('').str.replace(r'[\[\]"\']', '', regex=True)
data['genres_parsed'] = genres_raw.apply(lambda x: [g.strip() for g in x.split(',')] if x else [])
genres_exploded = data.explode('genres_parsed')
genre_counts = genres_exploded['genres_parsed'].value_counts().head(15)
plt.figure(figsize=(8, 6))
genre_counts.sort_values().plot(kind='barh', color='cadetblue')
plt.title('15 najczęściej występujących gatunków na Steam')
plt.xlabel('Ilość gier')
plt.ylabel('Gatunek')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Najczęściej występującym gatunkiem jest Indie, a następnie Action i Adventure – wskazuje to na dominację tych typów gier na platformie Steam. Są także kategorie mocno niszowe.

**5**

```python
plt.figure()
sns.scatterplot(data=data, x='price_initial (USD)', y='review_score', alpha=0.5)
plt.title("Cena gry a liczba recenzji")
plt.xlabel("Cena (USD)")
plt.ylabel("Liczba recenzji unormowana do 10")
plt.tight_layout()
plt.show()
```

Zarówno tanie, jak i drogie gry mogą mieć dużo lub mało recenzji.
Gry poniżej 100 USD pojawiają się w pełnym zakresie skali recenzji.
Najwięcej punktów skupia się przy niskich cenach (0–30 USD).
Droższe gry (100–300 USD) mają umiarkowaną liczbę recenzji, ale jest ich mało.
Skrajnie droga gra (1000 USD) praktycznie nie ma recenzji.

```python
data['main_genre'] = data['genres'].fillna('').str.replace(r'[\[\]\'\"]', '', regex=True)
data['main_genre'] = data['main_genre'].apply(lambda x: x.split(',')[0].strip() if x else None)
top10 = data['main_genre'].value_counts().head(10).index
filtered = data[data['main_genre'].isin(top10)]
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=filtered,
    y='main_genre',
    x='price_initial (USD)',
    showfliers=False,
    color='cadetblue')
plt.xlabel('Cena gry')
plt.ylabel('Gatunek gry')
plt.title('Rozkład cen gier względem głównego gatunku')
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

Najdroższe średnio okazują się być gry Simulation. Gatunki takie jak RPG, Simulation i Racing mają szerszy rozrzut cen – ich pudełka są dłuższe, co oznacza, że zawierają gry zarówno tanie, jak i droższe.Casual i Violent mają mniejsze mediany i węższy rozrzut, co sugeruje, że gry w tych gatunkach są zazwyczaj tańsze i bardziej jednorodne cenowo.
